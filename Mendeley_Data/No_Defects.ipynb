```python
import pandas as pd
import numpy as np

excel_path = r'c:\Users\hexiongwu\Documents\VSCode_On_Windows\Qoder\Mendeley_Data\No_Defects.xlsx'
output_path = r'c:\Users\hexiongwu\Documents\VSCode_On_Windows\Qoder\Mendeley_Data\No_Defects.json'

xl = pd.ExcelFile(excel_path)

result = {}

for sheet_name in xl.sheet_names:
    df = pd.read_excel(xl, sheet_name=sheet_name, header=None)
    
    labels = df.iloc[0, :3].astype(str)
    units = df.iloc[1, :3].astype(str).str.replace('(', '').str.replace(')', '')
    
    combined_names = labels + '_' + units
    
    data = df.iloc[2:, :3]
    data.columns = combined_names
    data = data.reset_index(drop=True)
    
    for col in data.columns:
        data[col] = pd.to_numeric(data[col], errors='coerce')
    
    result[sheet_name] = data.to_dict(orient='records')

result_df = pd.DataFrame([result])
result_df.to_json(output_path, orient='records', indent=2)

print(f'JSON file saved to: {output_path}')
print(f'Total sheets processed: {len(xl.sheet_names)}')
print(f'Sample output for sheet "0_1" (first 2 records):')
print(result['0_1'][:2])
```
