# xlsx → json Conversion


```python
import json
import pandas as pd

# Paths
xlsx_path = "No_Defects.xlsx"
json_path = "No_Defects.json"

# Read all sheet names
sheets = pd.ExcelFile(xlsx_path, engine="openpyxl").sheet_names

result = {}
for sheet in sheets:
    df = pd.read_excel(xlsx_path, sheet_name=sheet, header=None, engine="openpyxl")

    # Build merged headers: "Name (Unit)" from row 0 (headers) and row 1 (units)
    headers = [f"{df.iloc[0, c]} {df.iloc[1, c]}" for c in range(3)]

    # Extract data rows (skip header + unit rows), first 3 columns only
    data = df.iloc[2:, :3].copy()
    data.columns = headers
    data = data.astype(float)

    result[sheet] = data.to_dict(orient="records")

# Write JSON
with open(json_path, "w", encoding="utf-8") as f:
    json.dump(result, f, indent=2, ensure_ascii=False)

print(f"Saved {len(result)} sheets to {json_path}")
```
